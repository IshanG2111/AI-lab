Numeric Operators

```python
a, b = 10, 3
print("Numeric operators with a=10, b=3:")
print("  a + b =", a + b)
print("  a - b =", a - b)
print("  a * b =", a * b)
print("  a / b =", a / b)    # true division -> float
print("  a // b =", a // b)  # floor division
print("  a % b =", a % b)    # remainder
print("  a ** b =", a ** b)  # power
```

TypeCasting

```python
num=123
num2=1
a=str(num+num2)
print(a)
```

Boolean and logical operators

```python
is_adult = age >= 18
print()
print("Boolean and comparisons:")
print("  age >= 18 ->", is_adult)
print("  male and is_adult ->", male and is_adult)
print("  not male ->", not male)
```

Operator precedence

```python
expr1 = 2 + 3 * 4           # multiplication before addition
expr2 = (2 + 3) * 4         # parentheses change order
print("Operator precedence:")
print("  2 + 3 * 4 =", expr1)
print("  (2 + 3) * 4 =", expr2)
```

String

```python
greeting = "Hello, " + name  # string concatenation using existing 'name'

str_literal = "Hello, Python" # Define str_literal here
repeat = (str_literal + " ") * 2

print()
print("String examples:")
print(" ", greeting)
print(" ", repeat)
```

Literal and Constants

```python
# Constants (by convention - uppercase names)
PI = 3.14159
MAX_USERS = 100

# Literals examples
int_literal = 42
float_literal = 3.14
str_literal = "Hello, Python"
bool_literal = True

print("Literals and constants:")
print(" ", int_literal, float_literal, str_literal, bool_literal)
print("  PI (constant):", PI, "MAX_USERS (constant):", MAX_USERS)
print()
```
