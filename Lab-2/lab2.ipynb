3. Complete the following tasks using Arrays -
I. Array Creation
II. Array Element Access
III. Array Length
IV. Looping Array Elements
V. Adding Array Elements using append()
VI. Removing array elements using pop()

```python
print("I. Array Creation")
arr = [10, 20, 30, 40, 50]
print(f"Array: {arr}")

print("\nII. Array Element Access")
print(f"First element: {arr[0]}")
print(f"Last element: {arr[-1]}")
print(f"Elements from index 1 to 3: {arr[1:3]}")

print("\nIII. Array Length")
array_length = len(arr)
print(f"Length of array: {array_length}")

print("\nIV. Looping Array Elements")
print("Using For loop:")
for item in arr:
    print(f"  {item}")

print("\nV. Adding Array Elements")
arr.append(60)
print(f"After adding 60: {arr}")

print("\nVI. Removing Array Elements")
removed_element = arr.pop()
print(f"Removed element: {removed_element}")
print(f"Array after removal: {arr}")
```

4. Complete the following tasks using Sets -
I. Set Creation and Set Length
II. Allocation of Different Data Types into a Set
III. Using set() constructor
IV. Accessing Set Items
V. Adding items to a Set using add() and update()
VI. Remove set item using remove (), pop() and discard()
VII. Use clear() and del() on a set
VIII. Set Looping - for and while
IX. Set Joining using union(), update(), intersection(), intersection_update()
X. Use difference(), difference_update(), isdisjoint() and symmetric_difference()

```python
print("I. Set Creation and Set Length")
fruits = {"apple", "orange", "pineapple", "banana"}
print(f"Fruits set: {fruits}")
print(f"Length: {len(fruits)}")

print("\nII. Allocation of Different Data Types into a Set")
mixed_set = {1, "hello", 3.14, True, (1, 2)}
print(f"Mixed set: {mixed_set}")

print("\nIII. Using set() constructor")
colors = set(["red", "green", "blue"])
print(f"Colors set: {colors}")

print("\nIV. Accessing Set Items")
print("Sets are unordered, so we access items by looping:")
for fruit in fruits:
    print(f"  {fruit}")
print(f"Is 'apple' in fruits? {'apple' in fruits}")

print("\nV. Adding items to a Set")
fruits.add("mango")
print(f"After add(): {fruits}")
fruits.update(["grape", "kiwi"])
print(f"After update(): {fruits}")

print("\nVI. Remove set item")
fruits.remove("kiwi")
print(f"After remove('kiwi'): {fruits}")
fruits.discard("grape")
print(f"After discard('grape'): {fruits}")
popped_item = fruits.pop()
print(f"Popped item: {popped_item}, Set: {fruits}")

print("\nVII. Use clear() and del()")
temp_set = {1, 2, 3}
temp_set.clear()
print(f"After clear(): {temp_set}")
del temp_set

print("\nVIII. Set Looping")
print("For loop:")
for item in fruits:
    print(f"  {item}")
print("While loop:")
items_list = list(fruits)
idx = 0
while idx < len(items_list):
    print(f"  {items_list[idx]}")
    idx += 1

print("\nIX. Set Joining")
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(f"Union: {set1.union(set2)}")
print(f"Intersection: {set1.intersection(set2)}")
set1_copy = set1.copy()
set1_copy.update(set2)
print(f"After update(): {set1_copy}")
set1.intersection_update(set2)
print(f"After intersection_update(): {set1}")

print("\nX. Use difference operations")
setA = {1, 2, 3, 4}
setB = {3, 4, 5, 6}
print(f"Difference (A-B): {setA.difference(setB)}")
print(f"Symmetric difference: {setA.symmetric_difference(setB)}")
print(f"Is disjoint? {setA.isdisjoint(setB)}")
setA_copy = setA.copy()
setA_copy.difference_update(setB)
print(f"After difference_update(): {setA_copy}")
```

5. Complete the following tasks using Dictionaries -
I. Dictionary Creation & Access using Key, and Updation
II. Dictionary Length, Type and dict() constructor
III. Dictionary Adding and Removing Items (pop(), popitem()) and Checking
IV. Using clear() and del() on Dictionary
V. Looping and Copying Dictionary using copy()
VI. Nested Dictionary - Creation, Accessing and Looping Items

```python
print("I. Dictionary Creation & Access using Key, and Updation")
student = {"name": "Ishan", "age": 21, "grade": "A"}
print(student)
print(student['name'])
student["age"] = 21
student["grade"] = "A+"
print(student)

print("\nII. Dictionary Length, Type and dict() constructor")
print(f"Length: {len(student)}")
print(type(student))
teacher = dict(name="Ayush", subject="Math", experience=3)
print(teacher)

print("\nIII. Dictionary Adding and Removing Items")
student["city"] = "Mumbai"
print(student)
removed = student.pop("grade")
print(f"Removed: {removed}")
print(student)
if "name" in student:
    print("name exists")

print("\nIV. Using clear() and del()")
temp = {"a": 1, "b": 2}
print(temp)
temp.clear()
print(temp)
sample = {"x": 10, "y": 20}
del sample["x"]
print(sample)

print("\nV. Looping and Copying Dictionary")
person = {"name": "JJ", "age": 20, "city": "Delhi"}
for key in person:
    print(f"{key}: {person[key]}")
person_copy = person.copy()
print(person_copy)

print("\nVI. Nested Dictionary")
family = {
    "father": {"name": "Aditya", "age": 45},
    "son": {"name": "Ishan", "age": 21}
}
print(family)
print(f"Father's name: {family['father']['name']}")
for member, info in family.items():
    print(f"{member}: {info}")
```
